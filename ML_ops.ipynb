more complex models.

components, it's tough to perform unit testing; thus, most testing

specific defects.

#### 3. Data Storage
Now this data is cleaned and useful for many purpose like dashboard or marketing or predictions etc.<br>
So these data are stored in data warehouses generally. But stored in SQL, NoSQL too by some companies.

**Data Warehouse** e.g., Snowflake, redshift, Google Bigquery

Together Data Ingestion, Data Transformation and Data Storage is known as ETL. 

#### 4. Data Pipeline integration
Now, we need to do ETL daily. So to automate this, Data pipeline integration is used. Orchestration tools are used for this. <br>
example:

* Task 1: Trigger data extraction from sales and inventory systems.
* Task 2: Run data cleaning and transformation scripts
* Task 3: Load the cleaned and transformed data into data warehouses.
* Task 4: Execute sql queries in the data warehouse to generate analytical reports.
* Task 5: Distribute reports via mail to the inventory management and marketing teams.

e.g., Apache Airflow is a powerful, open-source workflow orchestration tool that allows users to programmatically author, schedule, and monitor data pipelines and workflows using Python code and Directed Acyclic Graphs (DAGs)

#### 5. Data Validatoin
We check quality of incoming data in this from every source.

1. **Extraction phase**: Before extracting data from the source, we can check if data is available in expected format, file sizes, timestamps or presence of specific files or tables. These are initial validation checks.

2. **Transformation phase**: Bulk data validation checks occur here.
    - Data type check
    - Range check
    - Consistency check
    - Completedness check
    - Uniqueness check
    - Business Rule validation ( domain specific rules, like customer's credit limit must not exceed)

3. **Loading phase**: Before loading the transformed data into target systems (like in data warehouse), final validation ensure that data is ready for integration. 
    - Referential integrity check : Ensuring foreign key correctly link to primary key.
    - Data formatting check : Source and data formatting is same
    - Aggregation check : Verifying sum, avg etc. are correctly calculated and match expected values.

e.g., GX Core, pyspark data validation tool etc.

#### 6. Data Versioning
Data versioning is the process of tracking, storing, and managing different versions of datasets over time, similar to how Git tracks code changes. It ensures reproducibility, consistency, and collaboration in data-driven projects.

It is important because:
    - **Reproducibility**  Ensures experiments or ML models can be recreated with the same dataset.
    - **Collaboration**  Teams can track changes, revert to previous datasets, and avoid overwriting important data.
    - **Regulatory Compliance**  In industries like healthcare and finance, tracking data changes is required.
    - **Auditability**  Helps understand how data has evolved over time.

##### **Problems Without Data Versioning**
- You train a model with a dataset but dont version it.
- Later, when you retrain the model, results are different (because data changed).
- You cant go back to the exact dataset that produced the best results.
- If mistakes happen (e.g., accidental deletion), theres no way to restore the previous version.

In banks, let say if another day, new dataset is changed with some guidelines from RBI, then model prediction may go off, and it may lead to wrong credit churn prediction. When accuracy goes down in some time, much loss already happened. And without data versioning, we can't roll back to previous model that was working well. <br> Only option remains is new model based on this new data, and this will take time and until this time, so much loss will happen because many undeserved people will get loan.

**Tools used**: DVC, lakeFS, Delta Lake etc.

#### 8. Feature store
A feature store is a centralized, specialized data platform designed for machine learning, enabling the storage, management, and efficient retrieval of features, facilitating consistent and reliable data access for model training and inference. 

**Why?**
- Reusability
- Transfer of expertise
- Data quality
- Standardization

It is used when company runs 100s of ML models not for 3 or 4 models.

##### Challanges without feature store
- Redundant Work: Teams repeatedly engineer the same features, wasting time and resources.

- Inconsistent Features: Feature definitions may vary across teams, leading to inconsistent model performance.

- Training-Serving Skew: Without a centralized system, discrepancies between training and production features can degrade model accuracy.

- No reusability of features for different projects in future.

- Slower Development: Manual feature engineering slows down experimentation and deployment.

- Data Governance Issues: Lack of lineage, versioning, and auditability makes compliance and debugging difficult.

- Scalability Problems: Managing features manually becomes increasingly complex as datasets and teams grow.

Also feature versioning concept is there that if a feature is modified then we can also track changes when needed.

e.g., Google Feature Store, Qwak

4. Throughput : Throughput is a measure of how many units of information a system can process in a given amount of time. The metric is applied across systems and organizations to help determine the performance of computers and networks and monitor response times.

To prevent data leakage outside company.

1. Data encryption at rest, in pipelines transit and in use.
2. Role based access control
3. Data masking and anonymization
4. Audit Trails : Keeping detailed log of all the data access and changes to track how data is used and by whom.
5. Backup and disaster recovery

e.g., symantec data protection, delphix



#### 11. Data governance
Data governance is a set of policies and procedures that ensure data is secure, accurate, and used responsibly. It includes how data is collected, stored, processed, and disposed of. 

- Data governance helps ensure data integrity and data security by defining and implementing policies, standards and procedures for data collection, ownership, storage, processing and use.
- The data governance function helps ensure that verified data flows through secured pipelines to trusted endpoints and users.

1. Data Cataloguing : Helps organization organize and access their data assets.
2. Policy management : Allows the creation and enforcement of data governance policies.
3. Automated data lineage : Tracks the origin, movement and transformation of data.
4. Privacy and security : Help organizations comply with data protection regulations like GDPR.

e.g. SAP master data governance, Collibra
